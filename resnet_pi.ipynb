```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import os

pp_time=np.load(os.path.join('..', 'benchmark_results', 'models', 'resnet_model_50', 'pp_time.npy'))
p_time=np.load(os.path.join('..', 'benchmark_results', 'models', 'resnet_model_50', 'p_time.npy'))
t_time=np.load(os.path.join('..', 'benchmark_results', 'models', 'resnet_model_50', 't_time.npy'))

print('-------------FOR RPi Resnet 50 PB---------')
print('-------------For Inference-------------')
print(f'Total Number of Images : {len(p_time)}')
print(f'Max Inference Time is : {round(np.max(p_time[1:]), 6)} msecs')
print(f'Min Inference Time is : {round(np.min(p_time[1:]), 6)} msecs')
print(f'Average Inference Time is : {round(np.average(p_time[1:]), 6)} msecs')
print(f'Mean Inference Time is : {round(np.mean(p_time[1:]), 6)} msecs')

plt.plot(p_time[1:])
plt.ylabel('Time in secs')
plt.xlabel('Iterations')
plt.show()
```

```python
print('------------For Preprocessing----------')
print(f'Total Number of Images : {len(pp_time)}')
print(f'Max Preprocessing Time is : {round(np.max(pp_time[1:]), 6)} secs')
print(f'Min Preprocessing Time is : {round(np.min(pp_time[1:]), 6)} secs')
print(f'Average v Time is : {round(np.average(pp_time[1:]), 6)} secs')
print(f'Mean Preprocessing Time is : {round(np.mean(pp_time[1:]), 6)} secs')

plt.plot(pp_time)
plt.ylabel('Time in secs')
plt.xlabel('Iterations')
plt.show()
```

```python
print('----------Total Time-----------')
print(f'Total Number of Images : {len(t_time)}')
print(f'Max Time is : {round(np.max(t_time[1:]), 6)} msecs')
print(f'Min Time is : {round(np.min(t_time[1:]), 6)} secs')
print(f'Average Time is : {round(np.average(t_time[1:]), 6)} secs')
print(f'Average Time is : {round(np.mean(t_time[1:]), 6)} secs')

plt.plot(t_time)
plt.ylabel('Time in secs')
plt.xlabel('Iterations')
plt.show()
```
