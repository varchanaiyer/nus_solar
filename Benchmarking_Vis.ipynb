```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import os

pp_time=np.load(os.path.join('benchmarks', 'pp_time.npy'))
p_time=np.load(os.path.join('benchmarks', 'p_time.npy'))
t_time=np.load(os.path.join('benchmarks', 't_time.npy'))

print('-------------FOR MobileNet---------')
print('-------------FOR CPU---------')

print(f'Time taken to load model : 4.98 secs')

print('-------------For Inference-------------')
print(f'Total Number of Images : {len(p_time)}')
print(f'Max Inference Time is : {round(np.max(p_time[1:]), 6) * 1000} msecs')
print(f'Min Inference Time is : {round(np.min(p_time[1:]), 6) * 1000} msecs')
print(f'Average Inference Time is : {round(np.average(p_time[1:]), 6) * 1000} msecs')
print(f'Mean Inference Time is : {round(np.mean(p_time[1:]), 6) * 1000} msecs')

plt.plot(p_time[150:300])
plt.ylabel('Time in secs')
plt.xlabel('Iterations')
plt.show()
```

```python
print('------------For Preprocessing----------')
print(f'Total Number of Images : {len(pp_time)}')
print(f'Max Preprocessing Time is : {round(np.max(pp_time[1:]), 6) *1000} msecs')
print(f'Min Preprocessing Time is : {round(np.min(pp_time[1:]), 6) *1000} msecs')
print(f'Average v Time is : {round(np.average(pp_time[1:]), 6) *1000} msecs')
print(f'Mean Preprocessing Time is : {round(np.mean(pp_time[1:]), 6) * 1000} msecs')

plt.plot(pp_time[150:300])
plt.ylabel('Time in secs')
plt.xlabel('Iterations')
plt.show()
```

```python
print('----------Total Time-----------')
print(f'Total Number of Images : {len(t_time)}')
print(f'Max Time is : {round(np.max(t_time[1:]), 6) * 1000} msecs')
print(f'Min Time is : {round(np.min(t_time[1:]), 6) * 1000} msecs')
print(f'Average Time is : {round(np.average(t_time[1:]), 6) * 1000} msecs')
print(f'Average Time is : {round(np.mean(t_time[1:]), 6) * 1000} msecs')

plt.plot(t_time[150:300])
plt.ylabel('Time in secs')
plt.xlabel('Iterations')
plt.show()
```
